```python
df = pd.read_csv("Salary_Data.csv")
```

```python
df.corr()
```

```python
plt.figure(figsize = (8,5), facecolor= "LightBlue")
plt.scatter(df["YearsExperience"], df["Salary"])
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
```

```python
model1 = smf.ols("Salary ~ YearsExperience", data = df).fit()
model1.summary()
```

```python
model2 = smf.ols("Salary ~ np.log(YearsExperience)", data = df).fit()
model2.summary()
```

```python
model3 = smf.ols("np.log(Salary) ~ YearsExperience", data = df).fit()
model3.summary()
```

```python
df["Yrs_sq"] = df["YearsExperience"] ** 2
model4 = smf.ols("Salary ~ YearsExperience", data = df).fit()
model4.summary()
```

```python
plt.figure(figsize=(10,7),facecolor="lightblue")
plt.scatter(df.YearsExperience,df.Salary,color="black",label="actual");
plt.plot(df.YearsExperience,model1.predict(df["YearsExperience"]),label="predicted",color="red",linestyle="-.");
plt.scatter(df.YearsExperience,model1.predict(df["YearsExperience"]),color="red");

plt.xlabel("Waist");
plt.ylabel("AT")
```

```python
plt.figure(figsize=(10,7),facecolor="lightblue")
plt.scatter(df.YearsExperience,df.Salary,color="black",label="actual");
plt.plot(df.YearsExperience,model4.predict(df["YearsExperience"]),label="predicted",color="red",linestyle="-.");
plt.scatter(df.YearsExperience,model4.predict(df["YearsExperience"]),color="red");

plt.xlabel("Waist");
plt.ylabel("AT")
```
